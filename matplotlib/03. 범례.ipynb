## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

### 범례 설정
- https://matplotlib.org/stable/api/legend_api.html

```python
# 라이브러리 로드
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
x = [1, 2, 3]
y = [2, 4, 8]

# 범례(legend) 설정
g= plt.plot(x, y, label="데이터") # 1. label 설정
plt.legend() # 2. 띄워
```

- 두 개의 선을 하나의 차트에 그릴 때

```python
x1 = [0, 2, 7]
y1 = [2, 4, 8]

x2 = [0, 6, 9]
y2 = [1, 2, 10]


plt.plot(x1, y1, label="data1", color="#f72fb4")
plt.plot(x2, y2, label="data2", color="#38bdf1")
g= plt.legend(["data1", "data2"])
```

```python
x = [1, 2, 3]
y = [2, 4, 8]

plt.plot(x, y)
plt.legend(["데이터"], loc="right") # or (0.7, 0.3) 으로 좌표지정 가능 0 ~ 1 사이 값을 이용
plt.legend(["데이터"], loc=(0.1, 0.9)) 
```

><b>2) 범례 폰트 사이즈 지정</b>

```python
x1 = [0, 2, 7]
y1 = [2, 4, 8]

x2 = [0, 6, 9]
y2 = [1, 2, 10]


plt.plot(x1, y1, label="data1", color="#f72fb4")
plt.plot(x2, y2, label="data2", color="#38bdf1")
g= plt.legend(fontsize=15, ncol=2) 
# fontsize : legend 크기 설정
# ncol : legend 줄 당 개수
```

><b>3) 범례 행 수 지정</b>

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]

x2 = [1, 2, 3]
y2 = [1, 100, 100]
```

```python
plt.plot(x1,y1, color="blue")
plt.plot(x2,y2, color="red")
```
