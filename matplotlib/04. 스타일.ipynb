## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
# 라이브러리 로드
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
x = [1, 2, 3]
y = [2, 4, 8]

g= plt.plot(x, y)
```

#### 1. 선 굵기 : linewidth

```python
plt.plot(x, y, linewidth=10) # linewidth로 선 굵기 정함
```

#### 2. 마커
- https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html


```python
plt.plot(x, y, marker="o", markersize=50)
# 'o' : 동그라미
```

```python
# 마커 사이즈 변경

```

```python
# markeredgecolor : 마커 테두리 색상
# markerfacecolor : 마커 색상

g = plt.plot([4, 5, 6], marker='H')
g = plt.plot([3, 4, 5], marker='d')
g = plt.plot([2, 3, 4], marker='x')
g = plt.plot([1, 2, 3], marker=11)

```

```python
g = plt.plot([4, 5, 6], marker='$lol$', markeredgecolor="red", markerfacecolor="yellow")
```

#### 3. 선 모양 : linestyle(https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

```python

g = plt.plot([4, 5, 6], linestyle=':', linewidth=5)
```

```python
g = plt.plot([4, 5, 6], linestyle='--', linewidth=5)
```

```python
g = plt.plot([4, 5, 6], linestyle='-.', linewidth=5, color="b")
```

#### 4. 선 색상 : color

- https://matplotlib.org/stable/api/colors_api.html#module-matplotlib.colors

```python
g = plt.plot([4, 5, 6], linestyle='-.', linewidth=5, color="r")
```

```python
g = plt.plot([4, 5, 6], linestyle='-.', linewidth=5, color="g")
```

#### 5. 포맷 - fmt = '[marker][line][color]'

```python
g = plt.plot([4, 5, 6], "bv:")
```

```python
g = plt.plot([4, 5, 6], "go")
```

#### 6. 축약(https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
# mfc : markerfacecolor
# mec : markeredgecolor
# ls : linestyle

g = plt.plot([4, 5, 6], marker='$lol$', ls=":", mec="b", mfc="r", ms=10)

```

#### 7. 투명도

```python
g = plt.plot([4, 5, 6], marker='$lol$', ls=":", mec="b", mfc="r", ms=10, alpha=0.3)
```

#### 8. 그래프 크기

```python
plt.figure(figsize=(10, 5)) # weidth, height
g = plt.plot([4, 5, 6], marker='$♬$', ls=":", mec="b", mfc="r", ms=20)
```

```python
plt.figure(figsize=(10, 5), dpi=20) # dpi = 확대?
g = plt.plot([4, 5, 6], marker='$♬$', ls=":", mec="b", mfc="r", ms=20)
```

```python
g = plt.plot([4, 5, 6], marker='$lol$', ls=":", mec="b", mfc="r", ms=10)
```
