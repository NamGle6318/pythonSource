### 1) 막대차트(기본)

```python
# plt.plot() : 이건 선차트
# plt.bar() : 이건 막대차트

labels= ["강백호", "서태웅", "정대만"]
values = [175, 180.226664, 177]

plt.bar(labels, values)
```

><b>색상</b>

```python
plt.bar(labels, values, color="r")
```

```python
plt.bar(labels, values, color=["r", "g", "b"])
```

```python
plt.bar(labels, values, color=["r", "g", "b"], alpha=0.5)
```

><b>축 레이블 조정</b>

```python
plt.bar(labels, values)
plt.ylim(175, 190) # 리미트 시작, 끝 설정하여 175~190 사이로 보여줌
```

><b>막대 너비 조정</b>

```python
plt.figure(figsize=(5, 5)) # 가로 세로 막대 너비 조정
plt.bar(labels, values, width=0.5)
```

><b>축 레이블 rotation 설정</b>

```python
plt.bar(labels, values, width=0.5)
plt.xticks(rotation=45) # 틱 기울임 45도 조정
plt.yticks(rotation=45) # 틱 기울임 45도 조정
```

><b>막대 </b>

```python
plt.bar(labels, values)
g = plt.xticks(labels,['1번 학생','2번 학생','3번 학생', ]) # 원본, 바꿀거로 xticks 설정
```
