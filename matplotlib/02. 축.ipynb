## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
# 라이브러리 로드
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

### 타이틀 설정
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html#matplotlib.pyplot.title

```python
# fontdict : 폰트 관련 설정 가능
# loc : left, right, center 로 위치 정할 수 있음


plt.title("꺾은선 그래프", fontdict={"size":30}, loc="right")

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.show()
```

### 축설정
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html

#### 1. 이름 설정
><b>1) xlabel(), ylabel()</b>

```python



plt.title("꺾은 선 그래프", fontdict={"size":20})

# 축 제목
plt.xlabel("x 축", fontdict={"color":"red"})
plt.ylabel("y 축", fontdict={"color":"blue"})
g = plt.plot(x, y)
```

><b>2) 이름 및 다양한 설정</b>

matplotlib.pyplot.xlabel(xlabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)

- fontdict : font 설정을 dict 구조로 설정
- labelpad : default 4.0 설정됨(label과 차트 사이의 간격)
- **kwargs
    - color : rgb설정, 색상명 가능
- loc : 축 이름 위치 설정('left', 'center', 'right'),  y 축 (top, bottom, center)

```python
plt.title("꺾은선 그래프", fontdict={"size": 30}, loc="right")

fontdict1 = {"color": "red", "family": "D2Coding", "weight": "bold", "size": 14}
fontdict2 = {"color": "b", "weight": "bold", "size": 14}

# 축 제목
plt.xlabel("x 축", fontdict=fontdict1)
plt.ylabel("y 축", fontdict=fontdict2)
g = plt.plot(x, y)
```

#### 2. 축 눈금 설정 - xticks(), yticks()

```python
x = [1, 2, 3]
y = [2, 4, 8]


plt.xlabel("x 축", fontdict=fontdict1)
plt.ylabel("y 축", fontdict=fontdict2)
g = plt.plot(x, y)

# 틱 설정 - plt.plot 보다 이후에 해야함.
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
```
