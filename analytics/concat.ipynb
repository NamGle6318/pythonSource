```python
import pandas as pd
```

```python
# concat
# merge

df1 = pd.DataFrame(
    {
        "id": [1, 2, 3],
        "customer_id": [1, 2, 3],
        "customer_name": ["Robert", "Peter", "Dave"],
    }
)

df2 = pd.DataFrame(
    {
        "id": [1, 2, 4],
        "customer_id": [100, 200, 300],
        "order_date": ["2024-01-21", "2024-02-03", "2024-10-01"],
    }
)
```

```python
print(df1)
print(df2)
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1,df2], axis=1)
```

```python
# merge() : sql join 과 같은 개념
pd.merge(df1, df2, how='left', on='id')
```

```python
# merge() : sql join 과 같은 개념
pd.merge(df1, df2, how='right', on='id')
```

```python
# merge() : sql join 과 같은 개념
pd.merge(df1, df2, how='inner', on='id')
```

```python
# df1의 id column을 index로 사용하기
# df1.set_index("id", inplace=True)
df1

df2.set_index("id", inplace=True)
df2
```

```python
# index 기준으로 merge 하기
pd.merge(df1, df2, left_index=True, right_index=True)
```

```python
# index 기준으로 merge + outer join
pd.merge(df1, df2, how='outer', left_index=True, right_index=True)
```
