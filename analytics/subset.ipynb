```python
friend_dict_list = [
    {"name": "John", "age": 25, "job": "student"},
    {"name": "Nate", "age": 35, "job": "teacher"},
    {"name": "Jenny", "age": 30, "job": "developer"},
]
# 데이터프레임 생성
# 데이터프레임 : series 모아놓은 구조

friend_df = pd.DataFrame(friend_dict_list, index=["s1", "s2", "s3"])
friend_df
```

```python
df = pd.DataFrame({
    "name": ["Alfred", "Batman", "Catwoman"],
    "toy": [None, "Batmobile", "Bullwhip"],
    "born": [None,pd.Timestamp("19400425"), None], 
})
df
```

```python
df.dtypes # NaT = datetime 타입, Null

# None == 결측 데이터
# 데이터 분석시 결측 데이터를 포함하냐 안하냐에 따라 결과 개수가 차이나기 때문에 결정을 신중히 하도록 해야함.
```

```python
# 결측치 데이터 처리 방법
# 1. 제거
# 2. 임의의 값으로 채우기

# 제거 : dropna() : missing data를 날린다.
df.dropna(axis=1, how="all") 
```

```python
# fillna : null 값을 특정한 값으로 채움
df.fillna({"name":"noname", "toy":"Bat", "born":pd.Timestamp('2000-01-01')}) # 키 마다 변화 시킬 값을 지정 가능
```

```python
student_list = {
    "name": ["John", "Nate", "Edward", "Zara", "John"],
    "major": [
        "Compuster Science",
        "Compuster Science",
        "Compuster Science",
        "Paychology",
        "Paychology",
    ],
    "gender": [
        "male",
        "male",
        "male",
        "female",
        "male",
    ],
    "age": [
        40,
        35,
        None,
        10,
        None,
    ],
}

student_df = pd.DataFrame(student_list)
student_df
```

```python
student_df.groupby("major")["age"].median()  # Edward, John : major 같은 사람들의 평균값

```

```python
student_df["age"].fillna(student_df.groupby("major")["age"].median())
```

```python
student_df["age"].fillna(student_df.groupby("major")["age"].transform('median'), inplace=True) # major 칼럼을 기준으로 group하고 age의 평균 값을 null에 반영(원본 유지)
student_df
```
